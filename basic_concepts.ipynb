### Entropy

### KL Divergence

```python
pa = [0.1, 0.2, 0.3, 0.4]
pb = pa #[0.1, 0.7, 0.1, 0.1]
plt.plot(np.arange(0,len(pa), 1), pa)
plt.plot(np.arange(0, len(pb), 1), pb)
```

```python
stats.entropy(pa, pb) 
```

### Covariance

```python
import numpy as np
```

```python
mean = [0, 0]
cov1 = [[200, -100], [-100, 200]]

xy = np.random.multivariate_normal(mean, cov1, 5000)
x, y = xy.T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
cov2 = [[10, 0], [0, 10]]

xy = np.random.multivariate_normal(mean, cov2, 5000)
x, y = xy.T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
xy = xy.T
```

```python
xy = np.matmul(np.array(cov1), xy)
x, y = xy
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
