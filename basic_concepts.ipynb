### Covariance

```python
import numpy as np
```

```python
mean = [0, 0]
cov1 = [[200, -100], [-100, 200]]

xy = np.random.multivariate_normal(mean, cov1, 5000)
x, y = xy.T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
cov2 = [[10, 0], [0, 10]]

xy = np.random.multivariate_normal(mean, cov2, 5000)
x, y = xy.T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
xy = xy.T
```

```python
xy = np.matmul(np.array(cov1), xy)
x, y = xy
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
## Metaclasses
```

```python
class Meta_1(type):
    def __new__(cls, name, bases, dct):
        x = super().__new__(cls, name, bases, dct)
        x.attr = 100
        return x    

    def __call__(cls, *a, **kw):
        print("entering Meta_1.__call__()")
        rv = super(Meta_1, cls).__call__(*a, **kw)
        print("exiting Meta_1.__call__()")
        return rv

class Class_1(metaclass=Meta_1):
    #__metaclass__ = Meta_1
    def __new__(cls, *a, **kw):
        print("entering Class_1.__new__()")
        rv = super(Class_1, cls).__new__(cls, *a, **kw)
        print("exiting Class_1.__new__()")
        return rv

    def __call__(self):
        print("Call of the object")

    def __init__(self, *a, **kw):
        print("executing Class_1.__init__()")
        super(Class_1,self).__init__(*a, **kw)
```
